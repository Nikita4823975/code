```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.covariance import EllipticEnvelope

true_cov = np.array([[0.6, 0.4], [0.4, 0.6]])
rng = np.random.default_rng(0)
X = rng.multivariate_normal(mean=[0, 0], cov=true_cov, size=500)

cov = EllipticEnvelope(random_state=0).fit(X)

new_points = np.array([[0.5, 1.2], [2, 1.1]])
predictions = cov.predict(new_points)
plt.figure(figsize=(8, 8))
plt.scatter(X[:, 0], X[:, 1], c='blue', alpha=0.5, label='Data Points')
plt.scatter(new_points[:, 0], new_points[:, 1], c='red', marker='x', s=100, label='New Points')

mean = cov.location_
covariance = cov.covariance_
eigenvalues, eigenvectors = np.linalg.eigh(covariance)
order = eigenvalues.argsort()[::-1]
eigenvalues, eigenvectors = eigenvalues[order], eigenvectors[:, order]
theta = np.linspace(0, 2 * np.pi, 100)
circle = np.c_[np.cos(theta), np.sin(theta)]

scale_factor = 2 
transform = scale_factor * np.sqrt(eigenvalues) * circle @ eigenvectors.T + mean

plt.plot(transform[:, 0], transform[:, 1], 'r-', lw=2, label='Elliptic Envelope')

plt.legend()
plt.title('Elliptic Envelope')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
cov.predict(new_points)
```
